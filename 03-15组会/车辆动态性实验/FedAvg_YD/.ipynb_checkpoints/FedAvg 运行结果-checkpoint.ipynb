```python
import matplotlib.pyplot as plt

import pandas as pd
import matplotlib.pyplot as plt

rr="/home/raoxy/FedProx-PyTorch/csv/"
# 读取10个CSV文件
csv_files =[
rr+'F1_car_1.csv',
rr+'F1_car_2.csv',
rr+'F1_car_3.csv',
rr+'F1_car_1.csv',
rr+'F1_car_2.csv',
rr+'F1_car_3.csv',
rr+'F1_car_1.csv',
rr+'F1_car_2.csv',
rr+'F1_car_3.csv',
rr+'F1_car_1.csv']

# 创建一个大图，包含10个子图
fig, axs = plt.subplots(2, 5, figsize=(15, 8))
# 逐个读取CSV文件并绘制子图
for i, csv_file in enumerate(csv_files):
    df = pd.read_csv(csv_file)
    ax = axs[i // 5, i % 5]
    ax.plot(df["accuracy"], label="Accuracy")
    ax.plot(df["precision"], label="Precision")
    ax.plot(df["recall"], label="Recall")
    ax.plot(df["f1"], label="F1 Score")
    ax.set_title(f"RSU {i+1}")
    ax.set_xlabel("round")  # 添加x轴标签
    if i in [0,5]:
        ax.set_ylabel("Y")  # 添加y轴标签
    ax.legend()

# 调整子图之间的间距
plt.tight_layout()
```
