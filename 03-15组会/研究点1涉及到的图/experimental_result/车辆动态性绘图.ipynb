```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
root = "/home/raoxy/experimental_result/yi_dong/"
# 定义每条折线的数据
y1 = pd.read_csv(root+"FedCan/FedCan100.csv")["loss"]
# 创建一个大图，并设置子图的布局为2x2
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

ax = axs[0, 0]
_ax = ax.twinx()
ax.plot(y1.index, pd.read_csv(root+"FedCan/FedCan100.csv")["loss"], label='FedCan',color="red",alpha=0.8)
ax.plot(y1.index, pd.read_csv(root+"FedProx/FedProx100.csv")["loss"], label='FedProx',color="green")
_ax.plot(y1.index, pd.read_csv(root+"FedAvg/FedAvg100.csv")["loss"], label='FedAvg', linestyle='--',color="blue",alpha=0.8)
ax.set_title('100%')
ax.set_xlabel("round", fontsize=12)
ax.set_ylabel("Val Loss")
ax.legend(loc='upper left')
_ax.legend(loc='upper right')
ax.grid()

ax2 = axs[0, 1]
_ax2 = ax2.twinx()
ax2.plot(y1.index, pd.read_csv(root+"FedCan/FedCan75.csv")["loss"], label='FedCan',color="red",alpha=0.8)
ax2.plot(y1.index, pd.read_csv(root+"FedProx/FedProx75.csv")["loss"],  label='FedProx',color="green")
_ax2.plot(y1.index, pd.read_csv(root+"FedAvg/FedAvg75.csv")["loss"], label='FedAvg', linestyle='--',color="blue",alpha=0.8)
ax2.set_title('75%')
ax2.set_xlabel("round", fontsize=12)
ax2.legend(loc='upper left')
_ax2.legend(loc='upper right')
ax2.grid()

ax3 = axs[1, 0]
_ax3 = ax3.twinx()
ax3.plot(y1.index, pd.read_csv(root+"FedCan/FedCan50.csv")["loss"], label='FedCan',color="red",alpha=0.8)
ax3.plot(y1.index, pd.read_csv(root+"FedProx/FedProx50.csv")["loss"],  label='FedProx',color="green")
_ax3.plot(y1.index, pd.read_csv(root+"FedAvg/FedAvg50.csv")["loss"], label='FedAvg', linestyle='--',color="blue",alpha=0.8)
ax3.set_title('50%')
ax3.set_xlabel("round", fontsize=12)
ax3.set_ylabel("Val Loss")
ax3.legend(loc='upper left')
_ax3.legend(loc='upper right')
ax3.grid()

ax4 = axs[1, 1]
_ax4 = ax4.twinx()
ax4.plot(y1.index, pd.read_csv(root+"FedCan/FedCan25.csv")["loss"], label='FedCan',color="red",alpha=0.8)
ax4.plot(y1.index, pd.read_csv(root+"FedProx/FedProx25.csv")["loss"],  label='FedProx',color="green")
_ax4.plot(y1.index, pd.read_csv(root+"FedAvg/FedAvg25.csv")["loss"], label='FedAvg', linestyle='--',color="blue",alpha=0.8)
ax4.set_title('25%')
ax4.set_xlabel("round", fontsize=12)
ax4.legend(loc='upper left')
_ax4.legend(loc='upper right')
ax4.grid()
# 调整子图之间的间距
plt.tight_layout()

# 显示图表
plt.show()
```
