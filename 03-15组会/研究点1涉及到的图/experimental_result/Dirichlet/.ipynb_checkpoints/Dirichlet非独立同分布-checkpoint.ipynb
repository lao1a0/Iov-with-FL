# alpha 1

````python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt

fed="/home/raoxy/experimental_result/Dirichlet/alpha1/"
# 读取10个CSV文件
csv_files =[ 'car_'+str(i)+'.csv' for i in range(1,11)]

# 创建一个大图，包含10个子图
fig, axs = plt.subplots(2, 5, figsize=(15, 8))
# 逐个读取CSV文件并绘制子图
for i, csv_file in enumerate(csv_files):
    ax = axs[i // 5, i % 5]
#     ax.plot(df["val_loss"])
#     ax2 = ax.twinx()
    ax.plot(pd.read_csv(fed+"FedCan/"+csv_file)["val_loss"],color="green",alpha=0.8,linewidth=2)
    ax.plot(pd.read_csv(fed+"FedAvg/"+csv_file)["val_loss"],color="blue",alpha=0.8,linestyle="--")
    ax.plot(pd.read_csv(fed+"FedProx/"+csv_file)["val_loss"],color="red")
    ax.set_title(f"RSU {i+1}")
    ax.set_xlabel("epoch")  # 添加x轴标签
    if i in [0,5]:
        ax.set_ylabel("Val Loss")  # 添加y轴标签
    ax.grid()
    
# 在大图的底部添加一个共同的图例
fig.legend(["FedCan", "FedAvg","FedProx"],
           loc="lower center", ncol=5, bbox_to_anchor=(0.5, -0.1),prop={'size': 14})

# 调整子图之间的间距
plt.tight_layout()
````

# 数据预处理

````python
import pandas as pd

for i in range(1,2):
    fn = '/home/raoxy/data/carhacking_dirichlet_compare/alpha05/car_{}.csv'.format(i)
    df = pd.read_csv(fn)
    df = df[df.columns[2:]]
    df[df["Label"]=="R"]=0
    df[df["Label"]=="RPM"]=1
    df[df["Label"]=="gear"]=2
    df[df["Label"]=="DoS"]=3
    df[df["Label"]=="Fuzzy"]=4
    df.to_csv(fn,index=False)
````
