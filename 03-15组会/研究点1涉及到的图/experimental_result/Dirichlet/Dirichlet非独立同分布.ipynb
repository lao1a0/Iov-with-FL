# alpha 1

````python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt

fed="/home/raoxy/experimental_result/Dirichlet/alpha1/"
# 10CSV
csv_files =[ 'car_'+str(i)+'.csv' for i in range(1,11)]

# 10
fig, axs = plt.subplots(2, 5, figsize=(15, 8))
# CSV
for i, csv_file in enumerate(csv_files):
    ax = axs[i // 5, i % 5]
#     ax.plot(df["val_loss"])
#     ax2 = ax.twinx()
    ax.plot(pd.read_csv(fed+"FedCan/"+csv_file)["val_loss"],color="red",alpha=0.8)
    ax.plot(pd.read_csv(fed+"FedAvg/"+csv_file)["val_loss"],color="blue",alpha=0.8,linestyle=":")
    ax.plot(pd.read_csv(fed+"FedProx/"+csv_file)["val_loss"],color="green")

    ax.set_title(f"RSU {i+1}")
    ax.set_xlabel("epoch")  # x
    if i in [0,5]:
        ax.set_ylabel("Val Loss")  # y
    ax.grid()
    
# 
fig.legend(["FedCan", "FedAvg","FedProx"],
           loc="lower center", ncol=5, bbox_to_anchor=(0.5, -0.1),prop={'size': 14})

# 
plt.tight_layout()
````
