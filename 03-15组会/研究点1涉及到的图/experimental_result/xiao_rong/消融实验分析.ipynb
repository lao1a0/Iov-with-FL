```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 读取数据
df1 = pd.read_csv("all have.csv")["loss"]
df2 = pd.read_csv("no_pre_procc.csv")["loss"]
# df3 = pd.read_csv("all no.csv")["loss"]
df3 = pd.read_csv("global_loss_acc.csv")["loss"]

# x坐标
x = np.arange(1, 51)

# 创建图表和坐标轴对象
fig, ax1 = plt.subplots(figsize=(8, 6))

# 绘制df1和df2的折线图，y轴在左边
ax1.plot(x, df1, label='All', color='blue', linestyle='-')
ax1.plot(x, df2, label='No Step 1', color='green', linestyle='--')

# 设置坐标轴标签
ax1.set_xlabel('rounds', fontsize=14)
ax1.set_ylabel('Test Loss', fontsize=14)

# 创建一个共享x轴的次坐标轴对象，用于df3
ax2 = ax1.twinx()
ax2.plot(x, df3, label='Base Line', color='red', linestyle=':', linewidth=2)

# 设置次坐标轴标签
ax2.set_ylabel('Test Loss (df3)', fontsize=14)

# 添加图例
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# 显示图表
plt.show()
```
