# 病态Non-IID

````python
import pandas as pd

# 读取10个CSV文件
csv_files =[ 'car_'+str(i)+'.csv' for i in range(1,11)]

# 创建一个大图，包含10个子图
fig, axs = plt.subplots(2, 5, figsize=(20, 8))
# 逐个读取CSV文件并绘制子图
for i, csv_file in enumerate(csv_files):
    ax = axs[i // 5, i % 5] 
    _ax2 = ax.twinx()
    ax.plot(pd.read_csv("experimental_result/EMD/CNN/csv/"+csv_file)["val_loss"],label="FedCan",color="red",alpha=0.8)
    _ax2.plot(pd.read_csv("experimental_result/bing_tai/FedAvg/"+csv_file)["val_loss"],label="FedAvg",color="blue",alpha=0.8,linestyle=":")
    ax.plot(pd.read_csv("experimental_result/bing_tai/FedProx/"+csv_file)["val_loss"],label="FedProx",color="green")
    ax.legend(loc="upper left")
    _ax2.legend(loc="upper right")
    ax.set_title(f"RSU {i+1}")
    ax.set_xlabel("epoch")  # 添加x轴标签
    if i in [0,5]:
        ax.set_ylabel("Val Loss")  # 添加y轴标签
    ax.grid()

    
# 在大图的底部添加一个共同的图例
# fig.legend(["FedCan", "FedProx","FedAvg"],
#            loc="lower center", ncol=5, bbox_to_anchor=(0.5, -0.1),prop={'size': 14})

# 调整子图之间的间距
plt.tight_layout()
````
