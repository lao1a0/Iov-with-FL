## CIFAR10 评价指标

```python
# 导入所需的库和模块
import torch
import torchvision
import torchvision.transforms as transforms
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix, roc_curve, auc

# 定义设备，可以是 CPU 或 GPU
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

# 加载MNIST数据集，并进行预处理
transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5,), (0.5,))])

testset = torchvision.datasets.CIFAR10(root='/home/raoxy/data', train=False,
                                       download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=4,
                                         shuffle=False, num_workers=2)

classes = [str(i) for i in range(1,11)]
criterion =  nn.CrossEntropyLoss().to(device)

# # 定义CNN网络的结构
class D_CNN(nn.Module):
    # 定义模型的构造函数
    def __init__(self):
        # 调用父类的构造函数
        super(D_CNN, self).__init__()
        # 定义激活函数为Sigmoid
        act = nn.Sigmoid
        # 定义模型的卷积部分，包括四个卷积层和两个池化层，以及Sigmoid激活函数
        self.body = nn.Sequential(
            # 第一个卷积层，输入通道为3，输出通道为12，卷积核大小为5，填充为2，步长为2
            nn.Conv2d(3, 12, kernel_size=5, padding=5//2, stride=2),
            # 第一个激活层，使用Sigmoid函数
            act(),
            # 第二个卷积层，输入通道为12，输出通道为12，卷积核大小为5，填充为2，步长为2
            nn.Conv2d(12, 12, kernel_size=5, padding=5//2, stride=2),
            # 第二个激活层，使用Sigmoid函数
            act(),
            # 第三个卷积层，输入通道为12，输出通道为12，卷积核大小为5，填充为2，步长为1
            nn.Conv2d(12, 12, kernel_size=5, padding=5//2, stride=1),
            # 第三个激活层，使用Sigmoid函数
            act(),
            # 第四个卷积层，输入通道为12，输出通道为12，卷积核大小为5，填充为2，步长为1
            nn.Conv2d(12, 12, kernel_size=5, padding=5//2, stride=1),
            # 第四个激活层，使用Sigmoid函数
            act(),
        )
        # 定义模型的全连接部分，包括一个线性层
        self.fc = nn.Sequential(
            # 第一个线性层，输入特征为768，输出特征为10，表示10个类别
            nn.Linear(768, 10),
            # 注释掉了第二个激活层和第二个线性层，可能是为了简化模型
            #act(),
            #nn.Linear(256, 10)
        )
        
    # 定义模型的前向传播函数
    def forward(self, x):
        # 将输入数据通过卷积部分，得到输出
        out = self.body(x)
        # 将输出展平为一维向量，形状为(N, 768)，其中N是批量大小
        feature = out.view(out.size(0), -1)
        # 打印特征的形状，用于调试
        #print(feature.size())
        # 将特征通过全连接部分，得到最终的输出
        out = self.fc(feature)
        # 返回输出和特征
        return out, feature
# # 加载训练好的 .pt 文件
```

```python
#### 模型替换

name='GC_CIFAR10.pth' #'Sotria_CIFAR10.pth'
model = D_CNN()
model.load_state_dict(torch.load('/home/raoxy/CIFAR10/iov-fl/'+name, map_location=device))
model.to(device)

loss_test=[]
acc_test=[]
y_true=[]
y_pred=[]
log_softmax = nn.LogSoftmax(dim=1)
def Pre(model, device, federated_test_loader, batch_size=200):
    global criterion,txt,loss_test,acc_test
    model.eval()
    test_loss = 0
    correct = 0
    n=0
    with torch.no_grad():
        for data, target in federated_test_loader:
#             model.send(data.location) # <-- NEW: send the model in virtual workers to Trusted Aggregator
            ##########################################################################
#             target = target.to(device) 
#             data=compress_channel(data, 28)
#             optimizer.zero_grad()
#             output = model(data.to(device))
            ##########################################################################
            data, target = data.to(device), target.to(device)
            output, feature_fc1_graph = model(data)
            output = log_softmax(output)
            #########################################################################
            n += target.shape[0]
            loss = criterion(output, target.long())
#             loss = F.nll_loss(output, target, reduction='sum')
#             model.get()
#             test_loss += loss.get() # sum up batch loss
            test_loss += loss.item() # sum up batch loss
            pred = output.argmax(1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
#             print("{}\t{}\t{}".format(pred.shape,target.shape,accuracy_score(pred.cpu(), target.cpu())))
            y_true.extend(target.cpu().numpy())
#             print(y_true)
            y_pred.extend(pred.squeeze().cpu().numpy())

    test_loss /= len(federated_test_loader)
    loss_test.append(test_loss)
    acc_test.append(correct*1.0/n)

    print('\tTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)'.format(
        test_loss, correct, len(federated_test_loader) * batch_size,
        100. * correct / n))
    return y_true,y_pred

y_true,y_pred=Pre(model, device, testloader, batch_size=200)

from sklearn.metrics import accuracy_score  # pip install scikit-learn
accuracy = accuracy_score(y_true, y_pred)
print("精度: ",accuracy)
```

```python
confusion = confusion_matrix(y_true, y_pred)
plt.imshow(confusion, cmap=plt.cm.Blues)
# plt.title("Confusion matrix")
plt.colorbar()
tick_marks = np.arange(10)
plt.xticks(tick_marks, classes)
plt.yticks(tick_marks, classes)
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.show()
```

```python
# 计算AUC值
y_true = np.array(y_true)
y_pred = np.array(y_pred)
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(10):
    fpr[i], tpr[i], _ = roc_curve(y_true == i, y_pred == i)
    roc_auc[i] = auc(fpr[i], tpr[i])

# 绘制ROC曲线
plt.figure()
for i in range(10):
    plt.plot(fpr[i], tpr[i], label='ROC curve of class {0} (area = {1:0.2f})'
                                   ''.format(i, roc_auc[i]))
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```

```python
# 导入需要的库
import pandas as pd
import matplotlib.pyplot as plt

# 定义要读取的文件名列表
files = [
    "DP-Gaussian_CIFAR10.csv",
    "GC_CIFAR10.csv",
    "DP-Laplace_CIFAR10.csv", 
    "Sotria_CIFAR10.csv",
    "My-CIFAR10.csv",
        ]

h='file/'

df1 = pd.read_csv(h+files[0],header=None, names=['DP-Gaussian_loss_train', 'DP-Gaussian_loss_test','DP-Gaussian_acc_train','DP-Gaussian_acc_test'])
df2 = pd.read_csv(h+files[1],header=None, names=['GC_loss_train', 'GC_loss_test','GC_acc_train','GC_acc_test'])
df3 = pd.read_csv(h+files[2],header=None, names=['DP-Laplace_loss_train', 'DP-Laplace_loss_test','DP-Laplace_acc_train','DP-Laplace_acc_test'])
df4 = pd.read_csv(h+files[3],header=None, names=['Sotria_loss_train', 'Sotria_loss_test','Sotria_acc_train','Sotria_acc_test'])
df5 = pd.read_csv(h+files[4],header=None, names=['My_loss_train', 'My_loss_test','My_acc_train','My_acc_test'])

df=pd.concat([df1,df2,df3,df4,df5], axis=1).iloc[1:]

df = df.astype(float) # 将数据框中的所有数据转换为浮点数
df.head()

# 将DataFrame中的数据进行可视化，设置两个y轴
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np

plt.rcParams['font.size'] = 16

# 定义要绘制的列名列表
cols_loss = ['DP-Gaussian_loss_test','DP-Laplace_loss_test','GC_loss_test','Sotria_loss_test','My_loss_test']
cols_acc = ['DP-Gaussian_acc_test','DP-Laplace_acc_test','GC_acc_test','Sotria_acc_test','My_acc_test']

# 定义要自定义的名称列表
names_loss = ['DP-Gaussian Loss','DP-Laplace Loss','GC Loss', 'Sotria Loss','SLGD Loss']
names_acc = ['DP-Gaussian Acc','DP-Laplace Loss','GC Acc','Sotria Acc', 'SLGD Acc']

# 定义要自定义的透明度列表
alphas_loss = [0.5, 0.5, 0.5,0.5, 1]
color_l=['#3c6158','#4b5cc4','#003472','#65abbc','#ff4777']
alphas_acc = [0.5, 0.5, 0.5,0.5, 1]
color_A=['#7d55c7','#00bc12', '#0c8918','#bed38e','#ff7500']
# 创建一个空的图形对象
fig = plt.figure(figsize=(8,6))

# 创建一个子图，位置为第一个，只有一行一列
ax = fig.add_subplot(1, 1, 1)
# 创建一个双轴对象，左轴为loss，右轴为acc
ax_loss = ax
ax_acc = ax.twinx()

# 设置左轴的纵轴标签为"Loss"
ax_loss.set_ylabel('Loss')
# 设置右轴的纵轴标签为"Accuracy"
ax_acc.set_ylabel('Accuracy')
# 设置横轴的标签为"Epoch"
ax.set_xlabel('Epoch')

# 遍历列名、名称和透明度列表，为每条折线绘制图形
for col_loss, name_loss, alpha_loss,color_l in zip(cols_loss, names_loss, alphas_loss,color_l):
    # 绘制loss折线，横轴为索引，纵轴为列名，标签为名称，颜色为蓝色，透明度为alpha_loss
    ax_loss.plot(df.index, df[col_loss], label=name_loss, color=color_l, alpha=alpha_loss)

j=0
for col_acc, name_acc, alpha_acc,color_A in zip(cols_acc, names_acc, alphas_acc,color_A):
    # 绘制acc折线，横轴为索引，纵轴为列名，标签为名称，颜色为橙色，透明度为alpha_acc
    ax_acc.plot(df.index, df[col_acc], label=name_acc, color=color_A, alpha=alpha_acc)

# 显示图例，位置为最佳
ax_acc.legend(loc='center left')
ax.legend(loc='center right')
# 显示网格
# ax.grid(True)
ax_acc.grid(True)
# 调整子图之间的间距
plt.tight_layout()
# 显示图形
plt.show()
```
