```python
import matplotlib.pyplot as plt

import pandas as pd
import matplotlib.pyplot as plt

rr="/home/raoxy/FedAvg_YD/csv/"
# 读取10个CSV文件
csv_files =[ rr+'F1_car_'+str(i)+'.csv' for i in range(1,11)]

# 创建一个大图，包含10个子图
fig, axs = plt.subplots(2, 5, figsize=(15, 8))
# 逐个读取CSV文件并绘制子图
for i, csv_file in enumerate(csv_files):
    df = pd.read_csv(csv_file)
    ax = axs[i // 5, i % 5]
    ax.plot(df["accuracy"])
    ax.plot(df["precision"],alpha=0.6)
    ax.plot(df["recall"],alpha=0.6)
    ax.plot(df["f1"],alpha=0.6)
    ax.set_title(f"Car {i+1}")
    ax.set_xlabel("round")  # 添加x轴标签
    if i in [0,5]:
        ax.set_ylabel("Y")  # 添加y轴标签
#     ax.legend()
    
# 在大图的底部添加一个共同的图例
fig.legend(["accuracy", "precision", "recall", "f1"],
           loc="lower center", ncol=5, bbox_to_anchor=(0.5, -0.1),prop={'size': 14})

# 调整子图之间的间距
plt.tight_layout()
```
