```python
def Psnr(image1,image2)
    # 计算psnr值
    psnr = skimage.metrics.peak_signal_noise_ratio(image1, image2, data_range=255)
    print('PSNR =', psnr)
```

```python
def SSIM(image1, image2):
    # 计算SSIM值
    ssim = skimage.metrics.structural_similarity(image1, image2, data_range=255,channel_axis=-1)
    print('SSIM =', ssim)
```

```python
def FMSE(image1,image2):
        # 导入需要的模块
    import skimage.io
    import sklearn.metrics
    import sklearn.feature_extraction.image

    # 提取图像的特征向量
    extractor = sklearn.feature_extraction.image.PatchExtractor(patch_size=(3, 3))
    feature1 = extractor.transform(image1.reshape(1, *image1.shape)).ravel()
    feature2 = extractor.transform(image2.reshape(1, *image2.shape)).ravel()

    # 计算FMSE值
    fmse = sklearn.metrics.mean_squared_error(feature1, feature2)
    print('FMSE =', fmse)
FMSE(image1,image2)
```

```python
def IPIPS(image1, image2):
    # 导入需要的模块
    import skimage.io
    import lpips

    # 读取两张图片
    image1 = skimage.io.imread('image1.jpg')
    image2 = skimage.io.imread('image2.jpg')

    # 创建一个LPIPS对象，使用'alex'网络
    lpips_alex = lpips.LPIPS(net='alex')

    # 计算LPIPS值
    lpips_value = lpips_alex(image1, image2, data_range=255)
    print('LPIPS =', lpips_value)
    
# IPIPS(image1, image2)
```

```python
def IIP_pixel(image1, image2):
    # 导入需要的模块
    import skimage.io
    import skimage.util
    import skimage.color
    import sklearn.metrics
    import numpy as np
    # 删除图片中的透明度通道
    image1_rgb = np.delete(image1, 3, axis=2)
    image2_rgb = np.delete(image2, 3, axis=2)

    # 反转第二张图片的像素
    image2_rgb_inverted = skimage.util.invert(image2_rgb)

    # 转换图片为灰度图片
    image1_gray = skimage.color.rgb2gray(image1_rgb)
    image2_gray = skimage.color.rgb2gray(image2_rgb)

    # 反转第二张图片的像素
    image2_gray_inverted = skimage.util.invert(image2_gray)
    # 计算IIP-pixel值
    iip_pixel = sklearn.metrics.mean_absolute_error(image1_gray, image2_gray_inverted)
    print('IIP-pixel =', iip_pixel)
    
IIP_pixel(image1, image2)
```
